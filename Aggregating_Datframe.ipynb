```python
import pandas as pd
import numpy as np
```

```python
station = pd.read_csv('weather_stations.csv')
weather = pd.read_csv('nyc_weather_2018.csv')
station.head()
```

```python
weather.head()
```

```python
station.rename(columns={"id" : "station"},inplace = True)
station.head()
```

```python
data = station.merge(weather, on = "station", how = "inner")
data.head()
```

```python
data = data.drop_duplicates(["station","datatype"]).reset_index(drop = True)
data.head()
```

```python
snow_avg = data.query('datatype == "SNOW" and name.str.contains("NJ US")',engine = "python")
print(snow_avg.elevation.mean())
```

```python
fb_data = pd.read_csv('fb_2018.csv')
fb_data.head()
```

```python
fb_changes = fb_data.assign(p_changes = lambda x: x.volume.pct_change())
fb_changes.head()
```

```python
fb_changes.sort_values(by = 'p_changes',ascending = False).head()
```

```python
earthquake = pd.read_csv("earthquakes.csv")
earthquake.head()
```

```python
data3 = earthquake[earthquake.magType == "ml"]
data3.head()
```

```python
earthquake = earthquake.assign(mag_bin = pd.cut(data3.mag,bins=5,labels=["nor","low","med","high","dangerous"]))
earthquake.mag_bin.value_counts()
```

```python
weather_by_station = pd.read_csv('weather_by_station.csv')
weather_by_station.head()
```

```python
weather_by_station = weather_by_station.assign(month = pd.DatetimeIndex(weather_by_station.date).month )
weather_by_station.head()
```

```python
sno = weather_by_station[weather_by_station.datatype == "SNOW"]
sno.reset_index(drop = True).head()
```

```python
a = pd.crosstab(index = sno.station, columns = sno.month, colnames=['month'], values = sno.value, aggfunc=np.mean)
a.head()
```
