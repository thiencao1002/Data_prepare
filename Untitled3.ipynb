```python
import pandas as pd
```

```python
df1 = pd.read_csv('aapl.csv',sep =',')
df2 = pd.read_csv('amzn.csv',sep = ',')
df3 = pd.read_csv('fb.csv',sep = ',')
df4 = pd.read_csv('goog.csv',sep = ',')
df5 = pd.read_csv('nflx.csv',sep = ',')
print(df1.shape)
print(df2.shape)
```

```python
df1[ 'ticker '] = 'AAPL'
df2[ 'ticker '] = 'AMZN'
df3[ 'ticker '] = 'FB'
df4[ 'ticker '] = 'GOOG'
df5[ 'ticker '] = 'NFLX'
```

```python
df = pd.concat([df1,df2,df3,df4,df5],axis = 0)
df.head()
```

```python
a = df.to_csv('faang.csv')
```

```python
newdf = df.assign(
    date = pd.to_datetime(df.date),
    volume = lambda x: x.volume.astype(int)
)

newdf.head()
```

```python
b = newdf.sort_values(by=['volume'],ascending=True)
b.head()
```

```python
df6 = pd.read_csv('covid19_cases.csv',sep = ',')
df6.head()
```

```python
new_df = df6.assign(
    date = pd.to_datetime(df6.dateRep)
)
new_df.head()
```

```python
new_df.set_index('date',inplace=True)
new_df.head()
```

```python
new_df.replace({'United_States_of_America':"US","United_ Kingdom":"UK"},inplace = True)
new_df.head()
```

```python
covid_clean = new_df.loc[new_df['countriesAndTerritories'].isin(["Argentina", "Brazil", "China", "Colombia", "India", "Italy", "Mexico", "Peru", "Russia", "Spain", "Turkey", "UK",  "USA"])]
covid_clean
```

```python
pivoted_df = new_df.pivot(index='dateRep',columns='countriesAndTerritories',values='cases').fillna(0)
pivoted_df
```
